````python
X_train, X_test = dados.iloc[:60_000, :], dados.iloc[60_000:, :]
y_train, y_test = resposta.iloc[:60_000].astype(np.int64), resposta.iloc[60_000:].astype(np.int64)
````

# Treinando um Classificador Binário

````python
X_train
````

````python
y_train_5 = y_train == 5
y_test_5 = y_test == 5
````

````python
y_train_5
````

````python
sgd_clf = SGDClassifier(random_state=42)
````

````python
sgd_clf.fit(X_train, y_train_5)
````

````python
y_train_5
````

````python
some_digit.shape
````

````python
sgd_clf.predict(some_digit.reshape(1, 784))
````

````python
acuracias = cross_val_score(sgd_clf, X_train, y_train_5, scoring='accuracy', cv=3)
````

````python
acuracias
````

````python
zero_r = DummyClassifier()
````

````python
zero_r.fit(X_train, y_train_5)
````

````python
acuracias_zero_r = cross_val_score(zero_r, X_train, y_train_5, scoring='accuracy', cv=3)
````
